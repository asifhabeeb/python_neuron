#### Problem: 2

Write a program which will find all such numbers which are divisible by 7 but are not a multiple
of 5, between 2000 and 3200 (both included). The numbers obtained should be printed in a
comma-separated sequence on a single line.

```python
numbers_list = list(range(2000, 3200, 1))
print([number for number in numbers_list if number % 7 == 0 and number % 5 != 0])
```

#### Problem: 4

Write a Python program to find the volume of a sphere with diameter 12 cm.
Formula: V=4/3 *  * r 3

```python
from math import pi

sphere_diameter = 12
sphere_volume = 4 / 3 * pi * (sphere_diameter / 2) ** 3
print("Volume of sphere: " + str(sphere_volume) + " cc")
```

#### Problem: 4

Write a Python Program to print the given string in the format specified in the sample output.
WE, THE PEOPLE OF INDIA, having solemnly resolved to constitute India into a
SOVEREIGN, SOCIALIST, SECULAR, DEMOCRATIC REPUBLIC and to secure to all
its citizens
Sample Output:
WE, THE PEOPLE OF INDIA,
having solemnly resolved to constitute India into a SOVEREIGN, !
SOCIALIST, SECULAR, DEMOCRATIC REPUBLIC
and to secure to all its citizens

```python
words = str('WE, THE PEOPLE OF INDIA, having solemnly resolved to constitute India into a '
            'SOVEREIGN, SOCIALIST, SECULAR, DEMOCRATIC REPUBLIC and to secure to all '
            'its citizens')

split_words = words.split(',')
counter = 0

for word in split_words:
    if word.isupper():
        print(word + ',', end='')
    else:
        further_split_words = word.split()
        if further_split_words[0] == 'DEMOCRATIC':  # TODO: Use a more general way to check this condition
            for next_word in further_split_words:
                if next_word.isupper():
                    print(' ' + next_word, end='')
                elif next_word.islower():
                    counter += 1
                    if counter == 1:
                        print("\n\t      ", end='')
                    print(next_word, end=' ')
            break
        print("\n      ", end='')
        for split_word in further_split_words:
            if split_word.isupper():
                print(split_word + ',!', end='')
                print("\n\t    ", end='')
            else:
                print(split_word, end=' ')
```
